```python
import json
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


def load_mlwd_json(json_path):

    with open(json_path, "r") as f:
        data = json.load(f)

    order = [
        "b1_s32_prefill",
        "b1_s64_prefill",
        "b1_s128_prefill",
        "b4_s32_prefill",
        "b4_s64_prefill",
        "b4_s128_prefill",
        "b1_s32_decode",
        "b1_s64_decode",
        "b1_s128_decode",
        "b4_s32_decode",
        "b4_s64_decode",
        "b4_s128_decode",
    ]

    matrix = []

    for key in order:
        row = [
            data[key]["sigma_bs"],
            data[key]["sigma_cu"],
            data[key]["sigma_l2"],
            data[key]["sigma_bw"],
        ]
        matrix.append(row)

    return np.array(matrix)


def plot_mlwd_heatmap(
        json1,
        json2,
        model1="Llama-3.2-3B-Instruct",
        model2="Qwen2.5-7B-Instruct",
        save_path=None
):

    data1 = load_mlwd_json(json1)
    data2 = load_mlwd_json(json2)

    x_labels = [
        "线程块调度器",
        "计算单元",
        "L2缓存",
        "显存带宽"
    ]

    y_labels = [
        "b=1,s=32,prefill",
        "b=1,s=64,prefill",
        "b=1,s=128,prefill",
        "b=4,s=32,prefill",
        "b=4,s=64,prefill",
        "b=4,s=128,prefill",
        "b=1,s=32,decode",
        "b=1,s=64,decode",
        "b=1,s=128,decode",
        "b=4,s=32,decode",
        "b=4,s=64,decode",
        "b=4,s=128,decode"
    ]

    vmin = min(data1.min(), data2.min())
    vmax = max(data1.max(), data2.max())

    plt.figure(figsize=(14,10))

    fig, axes = plt.subplots(
        1,2,
        figsize=(14,10),
        sharey=True
    )

    cmap = "RdBu_r"

    sns.heatmap(
        data1,
        ax=axes[0],
        cmap=cmap,
        vmin=vmin,
        vmax=vmax,
        center=1.0,
        annot=True,
        fmt=".2f",
        linewidths=0.5,
        xticklabels=x_labels,
        yticklabels=y_labels,
        cbar=False
    )

    axes[0].set_title(model1,fontsize=14)
    axes[0].hlines(6,*axes[0].get_xlim(),colors="black",linewidth=2)


    sns.heatmap(
        data2,
        ax=axes[1],
        cmap=cmap,
        vmin=vmin,
        vmax=vmax,
        center=1.0,
        annot=True,
        fmt=".2f",
        linewidths=0.5,
        xticklabels=x_labels,
        yticklabels=False,
        cbar=True,
        cbar_kws={
            "label":"干扰敏感度 (σ)"
        }
    )

    axes[1].set_title(model2,fontsize=14)
    axes[1].hlines(6,*axes[1].get_xlim(),colors="black",linewidth=2)


    plt.suptitle(
        "MLWD 干扰敏感度对比",
        fontsize=16
    )

    plt.tight_layout()

    if save_path:
        plt.savefig(
            save_path,
            dpi=300,
            bbox_inches="tight"
        )

    plt.show()
```
