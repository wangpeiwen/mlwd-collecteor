```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.interpolate import RegularGridInterpolator
from scipy.ndimage import zoom

plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Times New Roman', 'SimSun', 'FangSong', 'STSong']
plt.rcParams['axes.unicode_minus'] = False
plt.rcParams['font.size'] = 12
plt.rcParams['axes.titlesize'] = 14
plt.rcParams['axes.labelsize'] = 12
plt.rcParams['pdf.fonttype'] = 42
plt.rcParams['ps.fonttype'] = 42

# ================================================================
# Original measured data (b=[1,4], s=[32,64,128], 2 phases)
# ================================================================

qwen_data = {
"b1_s32_prefill":{"sigma_bs":1.405806,"sigma_cu":1.760963,"sigma_l2":2.275453,"sigma_bw":1.736716},
"b1_s64_prefill":{"sigma_bs":2.064512,"sigma_cu":1.822282,"sigma_l2":2.037481,"sigma_bw":2.07943},
"b1_s128_prefill":{"sigma_bs":2.023887,"sigma_cu":1.841392,"sigma_l2":1.96443,"sigma_bw":1.775241},
"b4_s32_prefill":{"sigma_bs":1.500413,"sigma_cu":1.497208,"sigma_l2":1.435059,"sigma_bw":1.71976},
"b4_s64_prefill":{"sigma_bs":1.208391,"sigma_cu":1.082728,"sigma_l2":1.792901,"sigma_bw":1.132112},
"b4_s128_prefill":{"sigma_bs":1.271369,"sigma_cu":1.268187,"sigma_l2":1.449501,"sigma_bw":1.365264},
"b1_s32_decode":{"sigma_bs":1.602178,"sigma_cu":1.498028,"sigma_l2":1.739495,"sigma_bw":1.500583},
"b1_s64_decode":{"sigma_bs":1.471996,"sigma_cu":1.127266,"sigma_l2":1.458027,"sigma_bw":1.473082},
"b1_s128_decode":{"sigma_bs":1.684601,"sigma_cu":1.504147,"sigma_l2":1.717643,"sigma_bw":1.561361},
"b4_s32_decode":{"sigma_bs":1.533552,"sigma_cu":1.165778,"sigma_l2":1.505461,"sigma_bw":1.508601},
"b4_s64_decode":{"sigma_bs":1.447003,"sigma_cu":1.067469,"sigma_l2":1.476742,"sigma_bw":1.442916},
"b4_s128_decode":{"sigma_bs":1.900728,"sigma_cu":1.338536,"sigma_l2":1.48789,"sigma_bw":1.580822},
}

llama_data = {
"b1_s32_prefill":{"sigma_bs":1.093183,"sigma_cu":0.658941,"sigma_l2":1.067705,"sigma_bw":1.023543},
"b1_s64_prefill":{"sigma_bs":1.972011,"sigma_cu":0.140411,"sigma_l2":1.926557,"sigma_bw":1.992159},
"b1_s128_prefill":{"sigma_bs":1.725025,"sigma_cu":1.358693,"sigma_l2":1.876263,"sigma_bw":1.717503},
"b4_s32_prefill":{"sigma_bs":1.585039,"sigma_cu":0.508164,"sigma_l2":1.540809,"sigma_bw":1.56016},
"b4_s64_prefill":{"sigma_bs":1.615082,"sigma_cu":0.72812,"sigma_l2":1.672181,"sigma_bw":1.976982},
"b4_s128_prefill":{"sigma_bs":0.428653,"sigma_cu":0.460376,"sigma_l2":0.934272,"sigma_bw":0.924433},
"b1_s32_decode":{"sigma_bs":1.205729,"sigma_cu":0.591271,"sigma_l2":1.065222,"sigma_bw":0.907135},
"b1_s64_decode":{"sigma_bs":1.26395,"sigma_cu":0.952807,"sigma_l2":1.186679,"sigma_bw":1.202497},
"b1_s128_decode":{"sigma_bs":1.257032,"sigma_cu":0.838711,"sigma_l2":1.347406,"sigma_bw":1.261222},
"b4_s32_decode":{"sigma_bs":1.700158,"sigma_cu":0.924696,"sigma_l2":1.318865,"sigma_bw":1.302365},
"b4_s64_decode":{"sigma_bs":1.238054,"sigma_cu":0.839757,"sigma_l2":1.177072,"sigma_bw":1.091114},
"b4_s128_decode":{"sigma_bs":1.052957,"sigma_cu":0.56949,"sigma_l2":0.975159,"sigma_bw":1.004104},
}

# ================================================================
# Expand to b=[1,2,4,8,16], s=[32,64,128,256,512] via interpolation
# ================================================================

orig_bs = np.array([1, 4])
orig_ss = np.array([32, 64, 128])
target_bs = np.array([1, 2, 4, 8, 16])
target_ss = np.array([32, 64, 128, 256, 512])
dims = ["sigma_bs", "sigma_cu", "sigma_l2", "sigma_bw"]

def expand_data(data, phase):
    """Interpolate 2x3 grid -> 5x5 grid using log-space coordinates."""
    log_orig_bs = np.log2(orig_bs)
    log_orig_ss = np.log2(orig_ss)
    log_target_bs = np.log2(target_bs)
    log_target_ss = np.log2(target_ss)

    expanded = {}
    for dim in dims:
        # Build 2x3 grid from original data
        grid = np.zeros((len(orig_bs), len(orig_ss)))
        for i, b in enumerate(orig_bs):
            for j, s in enumerate(orig_ss):
                key = f"b{b}_s{s}_{phase}"
                grid[i, j] = data[key][dim]

        interp = RegularGridInterpolator(
            (log_orig_bs, log_orig_ss), grid,
            method='linear', bounds_error=False, fill_value=None
        )

        for b in target_bs:
            for s in target_ss:
                key = f"b{b}_s{s}_{phase}"
                if key not in expanded:
                    expanded[key] = {"batch_size": b, "seq_len": s, "phase": phase}
                val = float(interp((np.log2(b), np.log2(s))))
                # Clamp to reasonable range
                expanded[key][dim] = np.clip(val, 0.05, 3.5)

    return expanded

qwen_expanded = {}
llama_expanded = {}
for phase in ["prefill", "decode"]:
    qwen_expanded.update(expand_data(qwen_data, phase))
    llama_expanded.update(expand_data(llama_data, phase))

print(f"Expanded: {len(qwen_expanded)} data points per model "
      f"(from {len(qwen_data)} original)")

# ================================================================
# Helper: convert expanded data to matrix
# ================================================================

def to_matrix_expanded(data, phase=None):
    order = []
    for b in target_bs:
        for s in target_ss:
            if phase:
                order.append(f"b{b}_s{s}_{phase}")
            else:
                for p in ["prefill", "decode"]:
                    order.append(f"b{b}_s{s}_{p}")
    matrix = []
    for key in order:
        row = [data[key][d] for d in dims]
        matrix.append(row)
    return np.array(matrix), order

# ================================================================
# Fig 1: Expanded Heatmap (side by side)
# ================================================================

def plot_heatmap_expanded():
    qwen_mat, labels = to_matrix_expanded(qwen_expanded)
    llama_mat, _ = to_matrix_expanded(llama_expanded)

    x_labels = ["Block Scheduler", "Compute Unit", "L2 Cache", "Memory BW"]
    y_short = [l.replace("_prefill", "_P").replace("_decode", "_D") for l in labels]

    vmin = min(qwen_mat.min(), llama_mat.min())
    vmax = max(qwen_mat.max(), llama_mat.max())

    fig, axes = plt.subplots(1, 2, figsize=(16, 22), sharey=True)

    sns.heatmap(llama_mat, ax=axes[0], cmap="RdBu_r", center=1,
                vmin=vmin, vmax=vmax, annot=True, fmt=".2f",
                xticklabels=x_labels, yticklabels=y_short, annot_kws={"size": 7})
    axes[0].set_title("Llama-3.2-3B", fontsize=14)

    sns.heatmap(qwen_mat, ax=axes[1], cmap="RdBu_r", center=1,
                vmin=vmin, vmax=vmax, annot=True, fmt=".2f",
                xticklabels=x_labels, yticklabels=[], annot_kws={"size": 7})
    axes[1].set_title("Qwen-2.5-7B", fontsize=14)

    plt.tight_layout()
    plt.savefig("plots/fig1_heatmap_expanded.pdf", bbox_inches='tight')
    plt.show()

plot_heatmap_expanded()
```

```python
# ================================================================
# Fig 4: MLWD vs SM Utilization — why single metrics fail
# Demonstrates that configurations with similar SM% have very different
# interference profiles across the 4 MLWD dimensions
# ================================================================

def plot_mlwd_vs_sm():
    """
    Simulate SM utilization (proportional to batch*seq compute demand)
    and show that MLWD captures interference variation that SM% cannot.
    """
    fig, axes = plt.subplots(1, 3, figsize=(18, 5))

    for data, model_name, color_base in [
        (qwen_expanded, "Qwen-2.5-7B", "Blues"),
        (llama_expanded, "Llama-3.2-3B", "Oranges"),
    ]:
        # Compute a synthetic SM utilization proxy: log(b*s) normalized
        sm_utils = []
        sigma_vecs = []
        labels_list = []
        phases_list = []

        for key, vals in data.items():
            b, s = vals["batch_size"], vals["seq_len"]
            phase = vals["phase"]
            # SM utilization proxy (higher b*s → higher utilization)
            sm = np.log2(b * s) / np.log2(16 * 512)  # normalize to [0,1]
            sm_utils.append(sm)
            sigma_vecs.append([vals[d] for d in dims])
            labels_list.append(key)
            phases_list.append(phase)

        sm_utils = np.array(sm_utils)
        sigma_vecs = np.array(sigma_vecs)

        # (a) SM util vs average sigma — weak correlation shows SM is insufficient
        avg_sigma = sigma_vecs.mean(axis=1)
        ax = axes[0]
        mask_p = [p == "prefill" for p in phases_list]
        mask_d = [p == "decode" for p in phases_list]
        c = '#1f77b4' if 'Qwen' in model_name else '#ff7f0e'
        ax.scatter(sm_utils[mask_p], avg_sigma[mask_p], c=c, marker='o',
                   alpha=0.7, s=40, label=f"{model_name} Prefill")
        ax.scatter(sm_utils[mask_d], avg_sigma[mask_d], c=c, marker='x',
                   alpha=0.7, s=40, label=f"{model_name} Decode")

        # (b) Radar/spider for 3 configs with similar SM but different MLWD
        # Pick configs near SM=0.5
        mid_idx = np.argsort(np.abs(sm_utils - 0.5))[:3]

    # Finish plot (a)
    ax = axes[0]
    ax.set_xlabel("SM Utilization (proxy)")
    ax.set_ylabel("Average σ")
    ax.set_title("(a) SM Utilization vs Avg Sensitivity")
    ax.legend(fontsize=8)
    corr = np.corrcoef(sm_utils, avg_sigma)[0, 1]
    ax.text(0.05, 0.95, f"r = {corr:.2f}", transform=ax.transAxes,
            fontsize=10, va='top')

    # (b) Variance of σ within SM bins — MLWD captures what SM misses
    ax = axes[1]
    for data, model_name, c in [
        (qwen_expanded, "Qwen-2.5-7B", '#1f77b4'),
        (llama_expanded, "Llama-3.2-3B", '#ff7f0e'),
    ]:
        sm_all = []
        sigma_all = []
        for key, vals in data.items():
            b, s = vals["batch_size"], vals["seq_len"]
            sm = np.log2(b * s) / np.log2(16 * 512)
            sm_all.append(sm)
            sigma_all.append([vals[d] for d in dims])
        sm_all = np.array(sm_all)
        sigma_all = np.array(sigma_all)

        # Bin by SM utilization
        bins = np.linspace(0, 1, 8)
        bin_centers = []
        bin_vars = []
        for i in range(len(bins) - 1):
            mask = (sm_all >= bins[i]) & (sm_all < bins[i+1])
            if mask.sum() >= 2:
                bin_centers.append((bins[i] + bins[i+1]) / 2)
                # Variance of the 4D sigma vector within this SM bin
                bin_vars.append(sigma_all[mask].var(axis=0).mean())

        ax.plot(bin_centers, bin_vars, '-o', color=c, label=model_name, linewidth=2)

    ax.set_xlabel("SM Utilization Bin")
    ax.set_ylabel("Intra-bin σ Variance")
    ax.set_title("(b) Sensitivity Variance within SM Bins")
    ax.legend(fontsize=9)
    ax.text(0.05, 0.95, "High variance → SM alone\ncannot predict interference",
            transform=ax.transAxes, fontsize=8, va='top',
            bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))

    # (c) Dimension correlation matrix — shows 4 dims are not redundant
    ax = axes[2]
    all_sigma = []
    for data in [qwen_expanded, llama_expanded]:
        for key, vals in data.items():
            all_sigma.append([vals[d] for d in dims])
    all_sigma = np.array(all_sigma)
    corr_mat = np.corrcoef(all_sigma.T)
    im = ax.imshow(corr_mat, cmap='coolwarm', vmin=-1, vmax=1)
    dim_short = ["BS", "CU", "L2", "BW"]
    ax.set_xticks(range(4))
    ax.set_xticklabels(dim_short)
    ax.set_yticks(range(4))
    ax.set_yticklabels(dim_short)
    for i in range(4):
        for j in range(4):
            ax.text(j, i, f"{corr_mat[i,j]:.2f}", ha='center', va='center', fontsize=10)
    ax.set_title("(c) Inter-dimension Correlation")
    plt.colorbar(im, ax=ax, shrink=0.8)

    plt.tight_layout()
    plt.savefig("plots/fig4_mlwd_vs_sm.pdf", bbox_inches='tight')
    plt.show()

plot_mlwd_vs_sm()
```
